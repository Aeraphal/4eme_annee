```python
import numpy as np
import cv2
from matplotlib import pyplot as plt



smarties=cv2.imread("smarties.png",cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(15,15))
plt.subplot(211)
plt.imshow(smarties,'gray')
```

```python
smarties2=abs(smarties-255)
[ret,thresh]= cv2.threshold(smarties2, 100, 255, cv2.THRESH_BINARY)
plt.figure(figsize=(15,15))
plt.imshow(thresh,'gray')
```

```python
S = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(6,4))
dist=cv2.distanceTransform(thresh,cv2.DIST_L2,3)
distinv=abs(255-dist)

smartieserode=cv2.erode(dist, S, iterations = 7)

plt.figure(figsize=(15,15))
plt.imshow(distinv,'gray')
```
